```python
import pandas as pd
import re

# Read the three CSV files into dataframes
try:
    facilitator_df = pd.read_csv('Facilitator_Responses.csv')
    peer_df = pd.read_csv('Peer_Responses.csv')
    no_ai_df = pd.read_csv('NoAI_Responses.csv')

    # Function to extract numbers from text like "Agree (5)" or "Disagree (2)"
    def extract_number(text):
        if isinstance(text, str):
            match = re.search(r'\((\d+)\)', text)
            if match:
                return int(match.group(1))
        return text # Return original value if not a string or no number in parentheses

    # Apply the function to all columns in each dataframe
    for df in [facilitator_df, peer_df, no_ai_df]:
        for col in df.columns:
            df[col] = df[col].apply(extract_number)

    print("Dataframes loaded and numbers extracted.")
    # Display the head of each dataframe to verify
    print("\nFacilitator DataFrame Head:")
    display(facilitator_df.head())
    print("\nPeer DataFrame Head:")
    display(peer_df.head())
    print("\nNoAI DataFrame Head:")
    display(no_ai_df.head())

except FileNotFoundError as e:
    print(f"Error loading file: {e}. Please make sure the CSV files are in the correct directory.")
except Exception as e:
    print(f"An error occurred: {e}")
```

## Perform non-parametric tests (for non-parametric data or when t-test assumptions are not met)

### Subtask:
Perform Kruskal-Wallis tests for each of the columns with indices 7-11 to compare the distributions across the Facilitator, Peer, and No AI conditions.


**Reasoning**:
Perform Kruskal-Wallis tests for columns 7-11 across the three conditions.



```python
from scipy.stats import kruskal

# Columns 7-11 were identified in a previous step
cols_7_11 = ['People on my team helped each other out when needed', 'We all cooperated to get the work done', 'On my team, people shared their knowledge with each other', 'On the whole, the members of this team all did their fair share of the work', 'My team coordinated activities to make this run smoothly']

# Iterate through the columns and perform Kruskal-Wallis test
for col in cols_7_11:
    # Extract data for each condition, dropping potential missing values
    facilitator_data = facilitator_df[col].dropna()
    peer_data = peer_df[col].dropna()
    no_ai_data = no_ai_df[col].dropna()

    # Perform Kruskal-Wallis test
    h_statistic, p_value = kruskal(facilitator_data, peer_data, no_ai_data)

    # Print the results
    print(f"Kruskal-Wallis test for: {col}")
    print(f"H-statistic: {h_statistic:.4f}, p-value: {p_value:.4f}")
    if p_value < 0.05:
        print("Interpretation: There is a statistically significant difference in the distribution across the three conditions.")
    else:
        print("Interpretation: There is no statistically significant difference in the distribution across the three conditions.")
    print("-" * 50)
```

## Perform non-parametric tests (for non-parametric data or when t-test assumptions are not met)

Perform Kruskal-Wallis tests for columns 12-17 and the average stress scores to compare distributions across the three conditions.



```python
from scipy.stats import kruskal

# Columns 12-17 were identified in a previous step
cols_12_17 = ['How Mentally demanding was the task?', 'How Physically demanding was the task?', 'How hurried or rushed was the pace of the task?', 'How successful were you in accomplishing what you were asked to do?', 'How hard did you have to work to accomplish your level of performance?', 'How insecure, discouraged, irritated, stressed and annoyed were you?']

# Iterate through the columns and perform Kruskal-Wallis test
for col in cols_12_17:
    # Extract data for each condition, dropping potential missing values
    facilitator_data = facilitator_df[col].dropna()
    peer_data = peer_df[col].dropna()
    no_ai_data = no_ai_df[col].dropna()

    # Perform Kruskal-Wallis test
    h_statistic, p_value = kruskal(facilitator_data, peer_data, no_ai_data)

    # Print the results
    print(f"Kruskal-Wallis test for: {col}")
    print(f"H-statistic: {h_statistic:.4f}, p-value: {p_value:.4f}")
    if p_value < 0.05:
        print("Interpretation: There is a statistically significant difference in the distribution across the three conditions.")
    else:
        print("Interpretation: There is no statistically significant difference in the distribution across the three conditions.")
    print("-" * 50)

# Perform Kruskal-Wallis test on the average stress scores
facilitator_avg_stress_data = combined_avg_stress_df[combined_avg_stress_df['Condition'] == 'Facilitator']['Average Stress Score'].dropna()
peer_avg_stress_data = combined_avg_stress_df[combined_avg_stress_df['Condition'] == 'Peer']['Average Stress Score'].dropna()
no_ai_avg_stress_data = combined_avg_stress_df[combined_avg_stress_df['Condition'] == 'No AI']['Average Stress Score'].dropna()

h_statistic_avg, p_value_avg = kruskal(facilitator_avg_stress_data, peer_avg_stress_data, no_ai_avg_stress_data)

print("\nKruskal-Wallis test for Average Stress Scores (Columns 12-17)")
print(f"H-statistic: {h_statistic_avg:.4f}, p-value: {p_value_avg:.4f}")
if p_value_avg < 0.05:
    print("Interpretation: There is a statistically significant difference in the distribution of average stress scores across the three conditions.")
else:
    print("Interpretation: There is no statistically significant difference in the distribution of average stress scores across the three conditions.")
```
